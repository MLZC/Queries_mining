```python
import codecs
import pickle
import jieba
import numpy as np
```

```python
%%time
# over view about Tencent_AILab_ChineseEmbedding
with codecs.open('./data/Tencent_AILab_ChineseEmbedding.txt', 'rb', encoding='utf-8') as f:
    for i, line in enumerate(f):
        segs = line.split()
        if i ==0:
            num = int(segs[0])
            dim = int(segs[1])
            print(num, dim)
            continue
        word = segs[0]
        vec = np.array(segs[1:], dtype=np.float64)
        print(word, vec)
        if i ==10:
            break
```

```python
%%time
with codecs.open('./data/Tencent_AILab_ChineseEmbedding.txt', 'rb', encoding='utf-8') as f:
    word2int_dic = {}
    int2word_dic = {}
    for i, line in enumerate(f):
        segs = line.split()
        if i ==0:
            num = segs[0]
            continue
        word = segs[0]
        if not word2int_dic.get(word,0):
            word2int_dic[word] = i
            int2word_dic[i] = word
        if i%1000000==0:
            print(word2int_dic[word])
            print(int2word_dic[i])
    print(len(word2int_dic))
    print(len(int2word_dic))
```

```python
# save python object as file
with open('./data/word2int_dic.pkl', 'wb') as f:
    pickle.dump(word2int_dic, f)
with open('./data/int2word_dic.pkl', 'wb') as f:
    pickle.dump(int2word_dic, f)
```

```python
%%time
# load python object from file
with open('./data/word2int_dic.pkl', 'rb') as f:
    word2int_dic = pickle.load(f)
with open('./data/int2word_dic.pkl', 'rb') as f:
    int2word_dic = pickle.load(f)
```

```python
%%time
def word_to_vector(n, file_name='./data/Tencent_AILab_ChineseEmbedding.txt'):
    '''
    file_name: path of word_embedding.txt
    n: the index of word in the word_embedding.txt
    '''
    with codecs.open(file_name, 'rb', encoding='utf-8') as f:
        for i, line in enumerate(f):
            if i < n:
                continue
            if i==n:
                segs = line.split()
                vec = np.array(segs[1:], dtype=np.float64)
                print(segs[0], vec, type(vec))
                break
word_to_vector(100000)
```
